### Baca File
Buka file, dan baca langsung seluruh konten dari file.

```python
f = open("months.txt")
print(f.read())
```

Buka file, dan baca per karakter

```python
f = open("months.txt")
nextChar = f.read(1)
while nextChar != "":
    print(nextChar)
    nextChar = f.read(1)
```

Baca per baris. 

```python
f = open("months.txt")
nextLine = f.readline()
while nextLine != "":
    print(nextLine)
    nextLine = f.readline()
```

Baca file per baris, dan strip(menghilangkan/membuant) karakter "\n" di akhir baris.

```python
f = open("months.txt")
nextLine = f.readline()
while nextLine != "":
    print( nextLine.strip() )
    nextLine = f.readline()
    
```

### Tokenization dengan split()


```python
text = "Ini bapak Budi. Bapak Budi pergi ke pasar. Di pasar ada ibu Budi. Bapak pulang ketemu Budi."
```

```python
text
```

```python
words = text.split()
```

```python
words
```

```python
sentences = text.lower().split(".")
sentences
```

```python
count = {}
for s in sentences:
    words = s.split()
    for w in words:
        if w not in count:
            count[w] = 1
        else:
            count[w] = count[w] + 1
```
